This file is test the performance of Anode with different levels of ODEs
$$Level\ 1:\ f(x,\theta)= \sigma (W_1X+b_1) $$

$$Level\ 2:\ f(x,\theta)= W_2\sigma (W_1X+b_1)+b_2 $$

$$Level\ 3:\ f(x,\theta)=W_3\sigma ( W_2\sigma (W_1X+b_1)+b_2 )+b_3$$


```python
import matplotlib.pyplot as plt
%matplotlib inline
import torch

device = torch.device('cpu')
```

## Create and visualize training data

```python
from experiments.dataloaders import ConcentricSphere
from torch.utils.data import DataLoader
from viz.plots import single_feature_plt

# Create training data in 2D, consisting of a concentric disk and an annulus
data_dim = 2
data_concentric = ConcentricSphere(data_dim, inner_range=(0., .5), outer_range=(1., 1.5), 
                                   num_points_inner=1000, num_points_outer=2000)
dataloader = DataLoader(data_concentric, batch_size=64, shuffle=True)

# Visualize a batch of data (use a large batch size for visualization)
dataloader_viz = DataLoader(data_concentric, batch_size=256, shuffle=True)
for inputs, targets in dataloader_viz:
    break

single_feature_plt(inputs, targets)
```

# Niveau 1

$$Level\ 1:\ f(x,\theta)= \sigma (W_1X+b_1) $$

Model training with augment dimension = 1

```python
from anode.models import ODENet
from anode.training import Trainer


hidden_dim = 32
num_epochs = 36

amodel = ODENet(device, data_dim, hidden_dim, time_dependent=True,augment_dim=1,
               non_linearity='relu',level=1,final_time=10.)

aoptimizer = torch.optim.Adam(amodel.parameters(), lr=1e-3)
atrainer = Trainer(amodel, aoptimizer, device,verbose=False)
atrainer.train(dataloader, num_epochs)
```

```python
from viz.plots import trajectory_plt

# To make the plot clearer, we will use a smaller batch of data
for small_inputs, small_targets in dataloader:
    break

trajectory_plt(amodel, small_inputs, small_targets, timesteps=10)
```

```python
from viz.plots import input_space_plt

input_space_plt(amodel)
```

```python
plt.plot(atrainer.histories['loss_history'])
plt.xlim(0, len(atrainer.histories['loss_history']) - 1)
plt.ylim(0)
plt.xlabel('Iterations')
plt.ylabel('Loss')
```

### Niveau 2
$$Level\ 2:\ f(x,\theta)= W_2\sigma (W_1X+b_1)+b_2 $$

Model training with augment dimension = 1

```python

amodel2 = ODENet(device, data_dim, hidden_dim, time_dependent=True,augment_dim=1,
               non_linearity='relu',level=2,final_time=10.)

aoptimizer2 = torch.optim.Adam(amodel2.parameters(), lr=1e-3)
atrainer2 = Trainer(amodel2, aoptimizer2, device)
atrainer2.train(dataloader, num_epochs)
```

```python
from viz.plots import trajectory_plt

# To make the plot clearer, we will use a smaller batch of data
for small_inputs, small_targets in dataloader:
    break

trajectory_plt(amodel2, small_inputs, small_targets, timesteps=10)
```

```python
from viz.plots import input_space_plt

input_space_plt(amodel2)
```

```python
plt.plot(atrainer.histories['loss_history'],label='ANODE Level 1')
plt.plot(atrainer2.histories['loss_history'],color='red',label='ANODE Level 2')
plt.xlim(0, len(atrainer.histories['loss_history']) - 1)
plt.ylim(0,1)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
```

### Origin Level 3
$$Level\ 3:\ f(x,\theta)=W_3\sigma ( W_2\sigma (W_1X+b_1)+b_2 )+b_3$$

Model training with augment dimension = 1

```python
amodel0 = ODENet(device, data_dim, hidden_dim, time_dependent=True,augment_dim=1,
               non_linearity='relu',final_time=10.)

aoptimizer0 = torch.optim.Adam(amodel0.parameters(), lr=1e-3)
atrainer0 = Trainer(amodel0, aoptimizer0, device)
atrainer0.train(dataloader, num_epochs)
```

```python
from viz.plots import trajectory_plt

# To make the plot clearer, we will use a smaller batch of data
for small_inputs, small_targets in dataloader:
    break

trajectory_plt(amodel0, small_inputs, small_targets, timesteps=10)
```

```python
from viz.plots import input_space_plt

input_space_plt(amodel0)
```

```python
plt.plot(atrainer.histories['loss_history'],label='ANode Niveau1')
plt.plot(atrainer2.histories['loss_history'],color='red',label='ANODE Niveau2')
plt.plot(atrainer0.histories['loss_history'],color='black',label='ANODE Origin')
plt.xlim(0, len(atrainer.histories['loss_history']) - 1)
plt.ylim(0,1)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
```
