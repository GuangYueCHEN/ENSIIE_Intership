
This file compare NODE, ANODE, Hamiltonian Inspired ODE and Hamiltonian ODE Network(to do).

```python
import matplotlib.pyplot as plt
import numpy
import torch
import torch.nn as nn
from torch.utils.data import DataLoader,TensorDataset
%matplotlib inline
from sklearn.datasets import make_blobs
dataset,lable= make_blobs(n_samples=3000, n_features=2, centers=5,random_state=10)
data=TensorDataset(torch.tensor(dataset).float(),torch.tensor(lable))
dataloader = DataLoader(data, batch_size=64, shuffle=True)
dataloader_viz = DataLoader(data, batch_size=256, shuffle=True)

for inputs, targets in dataloader_viz:
    break
   
plt.scatter(inputs[:, 0], inputs[:, 1], c=targets)
plt.show()

import torch.nn as nn

device = torch.device('cpu')
from anode.models import ODENet
from anode.training import Trainer

hidden_dim = 32
num_epochs = 24
data_dim =2 
from viz.plots import trajectory_plt

# To make the plot clearer, we will use a smaller batch of data
for small_inputs, small_targets in dataloader:
    break

from viz.plots import input_space_plt
from viz.plots import trajectory_plt2

print(targets.shape)
print(inputs.shape)
```

# NODE

```python
num_epochs = 100
model = ODENet(device, data_dim, hidden_dim, output_dim=5, time_dependent=False,
               non_linearity='relu')

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3 )

trainer = Trainer(model, optimizer, device,verbose=False,classification=True)
trainer.train(dataloader, num_epochs=5)
```

```python

trajectory_plt2(model, small_inputs, small_targets, timesteps=10)
```

```python
plt.plot(trainer.histories['loss_history'])
plt.xlim(0, len(trainer.histories['loss_history']) - 1)
plt.ylim(0)
plt.xlabel('Iterations')
plt.ylabel('Loss')
```

```python

res=nn.functional.log_softmax(model(inputs),dim=1).detach().numpy()

#small_inputs, small_targets
res=numpy.argmin(-res, axis=1)
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
plt.scatter(inputs[:, 0], inputs[:, 1], c= targets)
plt.subplot(1, 2, 2)
plt.scatter(inputs[:, 0], inputs[:, 1], c=res)
plt.show()
```

# Anode

```python

amodel = ODENet(device, data_dim, hidden_dim, output_dim=5, time_dependent=False,augment_dim=5,
               non_linearity='relu')

aoptimizer = torch.optim.Adam(amodel.parameters(), lr=1e-3)

atrainer = Trainer(amodel, aoptimizer, device,verbose=False,classification=True)
atrainer.train(dataloader, num_epochs=5)
```

```python
from viz.plots import trajectory_plt2
trajectory_plt2(amodel, small_inputs, small_targets, timesteps=10)
```
